### Views

#### Character Interactions

- ```Characters``` interacted if they appeared in the same ```scene.```


- Create the following view:
    - ```characterName_1```
    - ```character_id_1```
    - ```characterName_2```
    - ```character_id_2```
    - ```scene_count,``` number of scenes containing both ```characters.```
    - Order by ```scene_count``` descending.

- Each pair should appear only once, that is the table should not contain two rows of the form:
    - ```character_1, character_2, scene_count```
    - ```character_2, character_1, scene_count```
    
    

   - Create a temporary table temp which left joins scenes_characters with characters on characterName. This ensures that table temp has all the characterNames associated with their corresponding character_ids.

````python
%%sql
DROP TABLE IF EXISTS temp;
CREATE TABLE temp AS 
SELECT scenes_characters.*, characters.character_id
FROM scenes_characters 
LEFT JOIN characters 
ON scenes_characters.characterName = characters.characterName
````

````python
%sql SELECT * FROM temp LIMIT 10
````

   - Since there exists some characters with multiple character_ids (eg. White Walker: CH_362, CH_363, CH_364), and we cannot find out which character id correspondes to the character name in scenes_characters table, I choose to concatenate the character ids for one character name if there exists multiple ids for him/her.

````python
%%sql
DROP TABLE IF EXISTS temp1;
CREATE TABLE temp1 AS
SELECT DISTINCT seasonNum, episodeNum, sceneNo, characterName, GROUP_CONCAT(character_id) as character_id
FROM temp GROUP BY seasonNum, episodeNum, sceneNo, characterName
````

````python
%sql SELECT * FROM temp1 LIMIT 10
````

   - Create the character_interactions view.

````python
%%sql
DROP VIEW IF EXISTS character_interactions;
CREATE VIEW character_interactions AS
SELECT a.characterName AS characterName_1,
a.character_id AS character_id_1,
b.characterName AS characterName_2,
b.character_id AS character_id_2,
count(*) AS scene_count
FROM temp1 AS a
JOIN temp1 AS b
USING(seasonNum, episodeNum, sceneNo)
WHERE a.characterName < b.characterName
group by characterName_1, character_id_1, characterName_2, character_id_2
ORDER BY scene_count DESC
````

````python
%%sql SELECT * 
FROM character_interactions
LIMIT 30
````

#### Observation from character_interactions:
From the character_interactions view, we can find the count for two characters to be in the same scene, associated with the two characters' names and their ids. Thus, this view can show whether two characters have many interactions. If their scene_count is high, then they have many interactions.

#### Interesting View

- Create a view of the information that you think would be interesting to fans.


- Explain what you view is producing.

<u>Answer:</u>

   - I choose to create the view character_locations, which can show the relationship between one character and one location.

````python
%%sql
DROP VIEW IF EXISTS character_locations;
CREATE VIEW character_locations AS
SELECT a.location AS location,
b.characterName AS characterName,
count(*) AS location_count
FROM scenes AS a
JOIN scenes_characters AS b
USING(seasonNum, episodeNum, sceneNo)
group by location, characterName
ORDER BY location_count DESC
````

````python
%%sql SELECT * 
FROM character_locations
LIMIT 20
````

#### Observation from character_locations:
From this view, we can find the times of one character appearing in one location. Thus, this view can show the frequecy of one character being in one location.
